## Question 2: Finding Groupings in Human's Mitochondrial SNP/Mutations Patterns

This question focuses on mitochondria data extracted from the 1000 Genome Project dataset; a repository of 1024 individuals' genetic information displayed in terms of S "mutations".

A single-nucleotide polymorphism, often abbreviated to SNP (plural: snÉªps), is a variation in a single nucleotide that occurs at a specific position in the genome, where each variation is present to some appreciable degree within a population (e.g. > 1%).

For example, at a specific base position in the human genome, the C nucleotide may appear in most individuals, but in a minority of individuals, the position is occupied by an A. This means that there is a SNP at this specific position.

SNPs underlie differences in our susceptibility to disease; a wide range of human diseases, e.g. sickle-cell anemia and cystic fibrosis, result from SNPs. The severity of illness and the way the body responds to treatments are also manifestations of genetic variations. For example, a single-base mutation in the APOE (apolipoprotein E) gene is associated with a lower risk for Alzheimer's disease

A haplogroup is a group of individuals that share similar groups of genes together due to sharing a common ancestor. In human genetics, the haplogroups most commonly studied are Y-chromosome (Y-DNA) haplogroups and mitochondrial DNA (mtDNA) haplogroups, each of which can be used to define genetic populations.

### Importing necessary packages

```r
library(tidyverse)
library(factoextra)
library(cluster)
library(NbClust) 
library(fpc) 
library(dendroextras)
library(dendextend) 
library(mclust)  
library(dbscan)
library(dplyr)
```

### Replacing missing values with the median value

```r
for (i in 2:ncol(Mt)){
  Mt[is.na(Mt[,i]),i]<-median(Mt[,i],na.rm = TRUE)
}
```

### Deleting columns with 0 variance

```r
Mt_1=as.matrix(sapply(Mt[-1], as.numeric))
Mt_2<-as.data.frame(Mt_1[,apply(Mt_1,2,var,na.rm=TRUE) !=0])
Mt_2=cbind(Mt$Group,Mt_1)
colnames(Mt_2)[1]<-"Group"
dim(Mt_2)
```

## Principal Components Analysis (PCA)

```r
par(mfrow=c(1,2))
Mt.s=scale(Mt_2)
Mt.p=prcomp(Mt_2[,-1],scale=FALSE)
# Loading the principal components
Mt.p$rotation[1:5,1:5]
# Generating the rotation matrix for analysis
```

```r
summary(pvar)
```

```r
Mt.sd=Mt.p$sdev
Mt.var=Mt.p$sdev^2 
Mt.var[1:5]
# Computing standard deviation and variance for each principal component
pvar=Mt.var/sum(Mt.var) 
which.max(cumsum(pvar)[cumsum(pvar)<0.95])
which.max(cumsum(pvar)[cumsum(pvar)<0.98])
which.max(cumsum(pvar)[cumsum(pvar)<0.80])
# Calculating the number of PC's required to retain 80, 95 and 98 percent of the total variance
```

The first 3 PC's clearly indicate that the groupings are well defined and further analyses can be conducted successfully with them. However, the results indicate that 372 PC's have to be used to retain 95% of total variance and 555 PC's for retaining 98% of the total variance. This shows that most of the groupings (2712) are necessary for the analysis. 

From these results, we can see that the first 87 components account for 80% of the total variance.

```r
biplot(Mt.p)
```

## Clustering

```r
# First 100 principal components
cumsum(pvar[100])
fviz_screeplot(Mt.p,ncp=100,choice="eigenvalue")
plot(cumsum(pvar),xlab="Principal Component", 
   ylab="Cumulative Proportion of variance",ylim=c(0,1),type='b')
biplot(Mt.p,arrow.len=0) 
# Arrow head length is suppressed to get rid of the errors of indeterminate angle
Mtgroups1<- factor(Mt$Group)
plot(main="Groups",Mt.p$x[,1:100],col=Mtgroups)
Mt_new1=Mt.p$x[,1:100]
Mt_new1.s=scale(Mt_new1)
```

```r
# To generate the number of optimal clusters 
# Methods used (PC 1 - PC 100): NbClust, WSS, Gap statistic, and Silhouette
set.seed(10)
Mt.nbclust<-Mt_new1 %>% 
  scale() %>%
  NbClust(distance="euclidean",min.nc=2,max.nc=8,method="complete",index="all")
fviz_nbclust(Mt_new1,kmeans,method="wss") 
fviz_nbclust(Mt_new1,kmeans,method="gap_stat") 
fviz_nbclust(Mt_new1,kmeans,method="silhouette") 
```

The optimal number of clusters were calculated using NbClust, wss, gap statistic, and silhouette methods. Results revealed that:
NbClust: 2 clusters
wss: 8 clusters
gap statistic: 8 clusters
silhouette: 2 clusters
Therefore, the optimal number of clusters are 2 and 8 for 100 PC's

## Fuzzy K-means Clustering for K=2 and K=8 for 100 PC's

```r
set.seed(10)
kmeans100_2.fit=kmeans(Mt_new1,2,nstart=50)
attributes(kmeans100_2.fit)
kmeans100_2.fit$size
kmeans100_2.fit$tot.withinss
plotcluster(Mt_new1,kmeans100_2.fit$cluster)
fviz_cluster(kmeans100_2.fit,data=Mt_new1,ellipse.type="convex",palette="jco",ggtheme=theme_minimal())
```

```r
set.seed(9)
kmeans100_8.fit=kmeans(Mt_new1,8,nstart=50)
attributes(kmeans100_8.fit)
kmeans100_8.fit$size
kmeans100_8.fit$tot.withinss
plotcluster(Mt_new1,kmeans100_8.fit$cluster)
fviz_cluster(kmeans100_8.fit,data=Mt_new1,ellipse.type="convex",palette="jco",ggtheme=theme_minimal())
```

From these results, we can see that the total within sum of squares for k=2 is 15378.58 and for k=8 is 9637.87. Hence, k=8 is the optimal choice for clustering.

## Hierarchial clustering

```r
par(mfrow=c(1,2))
#Hierarchial Clustering with K=2
Mthc_ward=hclust(dist(Mt_new,method="euclidean"),method="ward.D2")
plot(Mthc_ward, main="Complete Linkage",xlab="",cex=.9) #Dendogram
rect.hclust(Mthc_ward,k=2,border="Red") # Bordering the clusters
```

```r
par(mfrow=c(1,2))
#Hierarchial Clustering with K=8
Mthc_ward=hclust(dist(Mt_new,method="euclidean"),method="ward.D2")
plot(Mthc_ward, main="Complete Linkage",xlab="",cex=.9) #Dendogram
rect.hclust(Mthc_ward,k=8,border="Brown") #Bordering the clusters
```

```r
# Segmenting for k=2 and k=8
clust_2=cutree(Mthc_ward,2)#Cut Tree into 2 clusters
clusplot(Mt_new1,clust_2,color=TRUE, shade = TRUE, labels=2, lines=0, main='Group segments')
clust_8=cutree(Mthc_ward,8)#Cut Tree into 8 clusters
clusplot(Mt_new1,clust_8,color=TRUE, shade = TRUE, labels=2, lines=0, main='Group segments')
```

```r
#Displaying the differences between clusters using discriminant coordinates
plotcluster(Mt_new1, clust_2) 
plotcluster(Mt_new1,clust_8)
```

The results of hierarchal clustering indicates that the chosen number of clusters (k=2 and k=8) are acceptable for analysis.
